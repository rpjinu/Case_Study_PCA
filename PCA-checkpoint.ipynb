# Principal Component Analysis

Now that we have understood correlation, lets first understand the need for PCA. 

You might ask the question, “How do I take all of the variables I’ve collected and focus on only a few of them?” In technical terms, you want to “reduce the dimension of your feature space.” By reducing the dimension of your feature space, you have fewer relationships between variables to consider and you are less likely to overfit your model. 

## Feature Extraction

The idea is that we are looking to reduce the dimensions of the original features set. But we don't reduce it by dropping out some of the columns, we rather create new columns from the existing ones. And we do this in such a way that we order our new features by their importance.

So lets supposed we create 5 new features from the dataset from original 32 features, these 5 new features are created from the original 5 and are created by ranking the original features in the order of importance. So we are not using any feature and we also get to work with less features!

Also since the new features are created from the older ones, even though we have less features now we are retaining their information, even with lesser features!

This process is called feature extraction. PCA is the technique for feature extraction.

Lets learn PCA and we will also be implementing PCA from scratch in python.

## Explaining PCA

PCA is a dimensionality reduction technique to reduce the number of features.  So if one is not able to identify the features or variables you want to remove from consideration and there are way too many variables to deal with. PCA is the way to go.
But how does it work? 

Lets see

Lets go through the steps one by one:
1. Standardise the data.


### Standardisation and why do it?
What is standardisation
Standardisation is a process of making sure a given feature has 0 mean and 1 variance, its one of the feature scaling techniques. Its achieved by the following formlua:

**Xstd= X-mean/std Dev**

Why do standardisation? Because PCA is a variance maximising algorithm, which means the features which show maximum variance are given more importance. Why again? Variance is a measure of how far a set of numbers is spread out from their average value. So more variance means that feature is going to add more value to the data. So we do need to look for features that are giving more variance. 
But if the features arent standardised, Some features may have more values in magnitude and hence PCA can try to give them more importance as they may show more  variance due to high magnitude and hence high magnitude changes and hence high variance.

Standardisation makes sure all the data features have 0 mean and 1 variance. 

So we standardise our original data and call this new standardised matrix as D.

```python
# standardize all the independent variables 
# we drop the target variable using the 'drop()'
# 'axis=1' drops the specified column
# the 'fit_transform()' transforms the data by first computing the mean and sd and later scaling the data
df_attr_std = StandardScaler().fit_transform(df.drop('Accident_severity', axis = 1))

# print the dimension of the standardized data
print(df_attr_std.shape)
```

2. Step 2 We compute the covariance matrix. Lets first understand covariance though.

## Covariance and Covariance Matrix
Covariance is the measure of how directional relationship of 2 variables. If the covariance is positive, it means that 2 variables move together or in the same direction, if it is negative that means they move in the opposite direction.
Covariance between two features is given by 

$ Cov(X,Y)=\sum \frac  {((X-\bar X)(Y- \bar Y)} {N-1} $

Covariance matrix is a DxD matrix, where D is the dimension of the features, so like we have 8 features in this diabetic dataset, our covariance matrix would be 8x8.


Covariance matrix gives us how each feature is related to the other i.e the relationship between the features, having the relationship between features is quite powerful in PCA, as ultimately we are trying to reduce the features.

Covariance matrix is computed as follows:
If D is the matrix of our features. Then 
**Cov(D) = DD<sup>T</sup>**

where D<sup>T</sup> is the transpose of D, tranpose is nothing but a transformation that change the rows to coliumn and vice versa of the given matrix.

```python
# generate the covariance matrix using 'cov' function
# consider the transpose of the dataframe 'scaledDf' to obtain the covarinace between the independent variables (columns)  
cov_mat = np.cov(df_attr_std.T)

# as 'cov_mat' is a numpy array, select first five observations with [0:5]
print(cov_mat[0:5])
```

3. Now that we have calculated the covariance matrix, lets focus on the next step, which is eigen value decomposition. Lets understand this step by step
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                  

## Eigen Values, Eigen vectors and Eigen Decomposition

Eigen vectors are special type of vectors that don't change their direction when a transformation is applied to the vector space, they only change the scalar value, and this scalar value is called Eigen value.
In more detailed fashion, if you have a 2D space and you transform it by applying a linear transformation, some vectors dont change their direction despite the transformation. Such vectors are called Eigen vectors, such vectors only change in scalar value, the value by which they change is called Eigen value.


The process of breaking down a matrix into its eigen vectors and eigen values is called eigen decomposition.
This is how Eigen decomposition is defined for a matrix Q.

$ \mathbf{A}=\mathbf{Q}\mathbf{\Lambda}\mathbf{Q}^{-1} $

Here Q is the square n × n matrix whose ith column is the eigenvector  and $ \Lambda $ is a diagonal matrix with the corresponding Eigen values.
So, every column in the Q correponds to the Eigen vector and every ith value in the diagonal of $ \Lambda $ corresponds to the eigen value of that eigen vector.
So for the first column of Q, ie. the first eigen vector, 1st value of the diagonal of $ \Lambda $ is going to be the corresponding eigen value.

We are going to be doing the Eigen decomposition of the covariance matrix in PCA.

Basically lets call the DD<sup>T</sup> matrix as Z, so we will be doing the eigen decomposition of Z.


```python
# use 'eig' function to compute eigenvalues and eigenvectors of the covariance matrix
eig_val, eig_vec = np.linalg.eig(cov_mat)

print('Eigenvalues:','\n','\n', eig_val,"\n")

print('Eigenvectors:','\n','\n',eig_vec,'\n')
```

**Interpretation:** For the 31x31 covariance matrix, we get 31 eigenvalues and eigenvectors. The eigenvector corresponding to the largest eigenvalue represent the direction of the highest variation in the dataset.

## Finding Principal components.

Finding principal components is just one more step ahead. 
4. We just get the principal components from the Eigen Decomposition.
       

To get the principal components, we need to use the eigen decomposition. So $ \Lambda $ is where the eigen values are, we just need to sort these eigen values in the decreasing order. Eigen values signifies the importance of a Eigen vector, so we also sort the Eigen vectors accordingly. For example if the second Eigen value lets call it $ \Lambda 2 $ has the maximum value, we change the 2nd column of the Eigen vector Q2 as the first vector and make it Q1.

Bottom line, we sort the Q matrix according to $ \Lambda $ 

Once sorted we get a new matrix called Q*, Q* is nothing but sorted version of Q, sorted according to $ \Lambda $ 
Now we just need to multiply our original matrix Z with Q*. So what we get is 
Z*= ZQ*
And there you have it, the columns of Z* are the principal components. These are new features of your data, ordered by importance. Z1 being the most important one.

_Bonus Knowledge, by definition principal components are the vectors perpendicular to each other, hence these new principal components are independent of each other._

Before performing this step in Python, lets first learn how to decide on how many principal components to consider

## Deciding on how many Principal components to take and scree plots

We are done understanding how PCA works, now there are 2 major ways of choosing how many of such newly built features to use? You can't use them all because too many features is why you chose to do PCA in the first place!

There are 3 ways to do this:
 

#### 1) Kaiser criterion: 
This criterion considers the number of pricipal components for which the eigenvalue is greater than 1. This criterion suffers a drawback of selecting more number of components as the eigenvalues very close to 1 may not contribute significantly in explaining the variation in the data.



```python
# create a list of eigenvalues
eig_val = list(eig_val)

# 'sort(reverse = True)' will sort the eigenvalues in the descending order
eig_val.sort(reverse = True)

# print the sorted list
print(eig_val)
```

Here the first 14 eigenvalues are greater than 1. Thus we can consider 14 principal components using kaiser criterion.

#### 2) Scree plot: 
Plot the eigenvalues and choose the number of components corresponding to the elbow point.

```python
# 'bp' represents blue color and pentagonal shape of points
plt.plot(eig_val,'bp')            

# plot a line plot
plt.plot(eig_val) 

# set the axes and plot labels
# set the fontsize using 'fontsize'
plt.title('Scree Plot', fontsize = 15)
plt.xlabel('Principal Components', fontsize = 15)    
plt.ylabel('Explained Variance', fontsize = 15)      

# use 'annonate' function to draw an arrow between points xy and xytext
# 's' is a argument to write text
# we can change the facecolor and arrowstyle; this is only for representation purpose
plt.annotate(text ='Elbow Point', xy=(2.65,1.2), xytext=(2, 1.8), arrowprops=dict(facecolor='black', arrowstyle = 'simple'))

# display the plot
plt.show()   
```

**Interpretation**: It can be observed that, after the elbow point, the principal components do not contribute much to the variance in the data. 
The Kaiser criterion considers the number of principal components as 15, but the scree plot shows that only first two components explains most of the variation.

#### 3) Percentage of Explained Variation: 
Calculate the percentage of variation explained by each principal component. We consider the number of components that explain most of the variation.

The eigenvalue represents the total variance explained by that principal component. 

The proportion of variation explained by the i<sup>th</sup> principal component is calculated by taking the ratio of i<sup>th</sup> eigenvalue to the sum of all the eigenvalues. To calculate the percentage, multiply this ratio by 100.

```python
# calculate the percentage of variation explained by each principal component
# create an empty list to store the percentage of variation explained by each principal component
percent_var = []

# use a 'for' loop to calculate the percentage of variation for each component
for i in eig_val:
    variation = (i/sum(eig_val))*100
    
    # add the values to the list 'percent_var'
    percent_var.append(variation)
    
# print the percentage of variation
percent_var
```

**Interpretation**: It can be seen that the first principal component explains 10.15% variation in the data.

Now calculate the cumulative percentage of variation.

```python
# the 'cumsum()' returns the cumulative sum 
np.cumsum(percent_var)
```

**Interpretation**: We can see that the first 6 components explain approximately 31% of the variation in the data. By considering the 2 components as per the scree plot, the components explain approximately 15% of the variation. To explain 95% of the variation in the data, we need to consider nearly all the principal components.

For this dataset, let us choose the `6` principal components that explain 31% of the variation.

### Finalizing PCA

```python
# consider the eigenvectors corresponding to the first five highest eigenvalues
# these eigenvectors are the 0,4,6,9,11,14 columns of 'eig_vec' 
eigenvector = eig_vec[:,[0,4,6,9,11,14]]

# print the vectors
eigenvector
```

# Types of PCA

## Randomised Vs Full PCA

When dataset is too large performing Eigen value decomposition could be a costly affair. When the m or n (number of training examples and the number of features) is greater that 500 and d (number of new features after reduction) in less that 80% of m or n, instead of full PCA scikit learn does randomised PCA. its a faster way of computing principal components.

Randomised PCA instead of directly computing eigen decomposition, it first computes a lower dimensional projection of the data using a randomised projection matrix. This randomised projection matrix is now used to find out principal components. This speeds up the computation pretty greatly!

In our case, scikit learn did Full PCA as data was less.
